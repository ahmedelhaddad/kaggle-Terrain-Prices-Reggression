```python
import pandas as pd
import numpy as np

df = pd.read_csv('train.csv')
df
```

```python
x=df.drop(['target', 'id', 'location_type', 'land_use', 'zoning_code'] , axis=1)
y=df['target']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler

# Apply Min-Max scaling to features (x)
scaler_x_minmax = MinMaxScaler()
x_scaled_minmax = scaler_x_minmax.fit_transform(x)

# Apply Min-Max scaling to target (y)
scaler_y_minmax = MinMaxScaler()
y_scaled_minmax = scaler_y_minmax.fit_transform(y.values.reshape(-1, 1)).flatten()

display(x_scaled_minmax[:5])
display(y_scaled_minmax[:5])
```

```python
x_scaled_minmax_df = pd.DataFrame(x_scaled_minmax, columns=x.columns)
y_scaled_minmax_df = pd.DataFrame(y_scaled_minmax, columns=['target'])

display(x_scaled_minmax_df)
display(y_scaled_minmax_df)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled_minmax_df, y_scaled_minmax_df, test_size=0.2, random_state=42)

print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model on the training data
model.fit(x_train, y_train)

print("Linear Regression model trained successfully.")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the test set
y_pred = model.predict(x_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2) score: {r2}")
```

```python
# Load the test data
test_df = pd.read_csv('test.csv')

# Preprocess the test data similar to the training data
x_test_data = test_df.drop(['id', 'location_type', 'land_use', 'zoning_code'], axis=1)

# Apply the same Min-Max scaling used for the training features
x_test_scaled_minmax = scaler_x_minmax.transform(x_test_data)

# Convert the scaled test features back to a DataFrame
x_test_scaled_minmax_df = pd.DataFrame(x_test_scaled_minmax, columns=x_test_data.columns)
```

```python
# Make predictions on the scaled test data
test_predictions_scaled = model.predict(x_test_scaled_minmax_df)

# Inverse transform the scaled predictions to get values in the original scale
test_predictions = scaler_y_minmax.inverse_transform(test_predictions_scaled.reshape(-1, 1)).flatten()
```

```python
# Create the submission DataFrame
submission_df = pd.DataFrame({'id': test_df['id'], 'target': test_predictions})

# Save the submission file
submission_df.to_csv('submission.csv', index=False)

print("Submission file created successfully!")
display(submission_df)
```
